```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("VideoGamesSales.csv")
```

```python
df2 = df[df.duplicated()]
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df["Region"] = df["Region"].fillna("North")
```

```python
df['Region'].isnull().shape
```

```python
df.head(100)
```

```python
df["NA_Sales"] = df["NA_Sales"].replace('[$]','', regex = True)
```

```python
df["NA_Sales"] = pd.to_numeric(df["NA_Sales"], errors = 'coerce')
```

```python
Average_Sales =  df["NA_Sales"].mean()
```

```python
Average_Sales
```

```python
Average_Sales  = int(Average_Sales)
```

```python
df["NA_Sales"] = df["NA_Sales"].fillna(Average_Sales) 
```

```python
df
```

```python
df["Country"] = df["Country"].replace({"USA":"United States"}) 
```

```python
df["Country"].str.title()
df2 = df[df["Country"] == "Australia"]
df = df.rename(columns = {"NA_Sales":"National Sales","Global_Sales":"Global Sales","NA_Profit":"National Profit","Global_Profit":"Global Profit"})
# sales_cap = 2000

# df["National Sales"] = np.where(sales_cap, df["National Sales"] > sales_cap,df["National Sales"])

df.head(100)
```

```python
# Create a Bar Chart(National Sales by Region and Country)
import seaborn as sns

National_Sales = df.groupby(["Region","Country"])["National Sales"].sum().reset_index()
plt.figure(figsize = (5,4))
sns.barplot(data = National_Sales, x = "Region", y = "National Sales", hue = "Country" )
plt.title("National Sales by Region and Country",fontweight = 'bold')
plt.xlabel("Region", fontweight = 'bold')
plt.ylabel("National Sales", fontweight = 'bold')
plt.xticks(rotation=45)

plt.grid(True)
plt.show()
```

```python
 # Create Box Plot('Global Sales By Platform)
plt.figure(figsize =(6,7))
sns.boxplot(x = 'Country',y = 'National Sales', 
            data  = df,
            showmeans =True,
            hue = 'Genre',
           meanprops = {"marker":"o",
                       "markerfacecolor":"white",
                       "markersize":"3",
                       "markeredgecolor":"green"}
           )

plt.title("Global Sales By Platform")
plt.xlabel("Country", fontweight = 'bold')
plt.ylabel('National Sales', fontweight = 'bold')
plt.xticks(rotation = 45)
plt.show()
```

```python
#  Create a pie chart
Sales = df.groupby("Country")[["National Sales","Global Sales"]].sum().reset_index()

fig, axs = plt.subplots(1,2, figsize =(10,4))

axs[0].pie(Sales["National Sales"],labels = Sales["Country"],autopct = "%1.1f%%",startangle = 140)
axs[0].set_title("Nation Sales", fontweight = 'bold')

axs[1].pie(Sales["Global Sales"], labels = Sales["Country"], autopct = "%1.1f%%")
axs[1].set_title("Global Sales by Country", fontweight = 'bold')


plt.show()
```
